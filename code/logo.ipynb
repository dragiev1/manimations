```python
from manim import *;
```

```python
class LogoAnimation(Scene):
    def construct(self):      
        logo = MathTex(
            r"\subset", "M", r"\sum",
            font_size=100,
            arg_separator=""  # Removes default spacing between args
        ) 
        logo[2].scale(0.5).shift(LEFT * 0.22 + UP * 0.04)
        logo[1].shift(DOWN * 0.05)

        logo.set_fill(opacity=0)
        logo.set_stroke(color=WHITE, width=2)
        logo.set_background_stroke(
            color = BLUE_E,
            width = 8,
            opacity = 0.4
        )
        
        # Center and animate
        logo.center().scale(1.5)
        self.play(Write(logo[0]), FadeIn(logo[1]), Write(logo[2]), run_time=2)
        self.play(Unwrite(logo[0]), FadeOut(logo[1], run_time = 1.5), Unwrite(logo[2]), run_time = 1.5, rate_func = linear)
        self.wait(0.5)
```

```python
manim -pqh LogoAnimation
```
