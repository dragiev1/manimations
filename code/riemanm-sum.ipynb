```python
from manim import *
```

```python
class RiemannSums(Scene):
  def construct(self):
    # Tracks how many rectangles are there
    tracker = ValueTracker(10)
    font_size = 32
    area_under_curve = Tex(R"Area under curve", font_size = font_size + 5)
    area_under_curve.to_corner(UP, buff=0.2).shift(LEFT * 2.75 + DOWN * 0.55).set_stroke(color=WHITE, width=1).set_fill(opacity=0)
    # Will hold the equations
    equation = VGroup()

    # Add axes 
    axes = Axes(
      y_range=(0, 5, 1),
      x_range=(0, 14, 1),
      x_length=14,
      y_length=8,
      axis_config={"include_tip": False}
    ) 
    axes.to_corner(DL, buff=0.2).scale(0.9)
    a_label = MathTex("a", font_size = 25)
    b_label = MathTex("b", font_size = 25)
    a_label.next_to(axes.x_axis.number_to_point(0), DOWN, buff=0.2)
    b_label.next_to(axes.x_axis.number_to_point(13), DOWN, buff=0.2)

    curve = axes.plot(
      lambda x: np.sin(x) + 3,
      x_range = (0, 13, 0.01),
      color = BLUE_E,
      stroke_width = 3
    )
    func = MathTex("f", font_size = 30).next_to(curve.get_end(), RIGHT)

    # Updates equations with new amount of rectangles 
    def update_eq(mob):
      n = int(tracker.get_value())
      new_eq = MathTex(
        Rf"\approx \sum_{{i=1}}^{{{n}}} f(c_i) \Delta x",
        font_size = font_size
      )
      # Position equations and scale + padding
      new_eq.next_to(area_under_curve, RIGHT).shift(DOWN * 0.05)
      mob.become(new_eq)

    # Attach the updater to equations to animate later
    equation.add_updater(update_eq)

    # Riemann rectangles to be displayed under curve
    riemann_rects = VGroup()

    # Updater for Riemann rects
    def update_rects(mob):
      n = int(tracker.get_value())
      rects = axes.get_riemann_rectangles(
        curve,
        x_range = (0, 13),
        dx = 8 / n,
        input_sample_type = "left",
        fill_opacity=0.3,
        color=[RED_A, RED_D]
      )
      mob.become(rects)
    riemann_rects.add_updater(update_rects)
    update_rects(riemann_rects)

    # Initialize starting value
    update_eq(equation)
    self.play(FadeIn(axes), run_time=0.5)
    self.play(Write(area_under_curve), Write(a_label), Write(b_label), run_time = 0.5)
    self.play(Write(equation), Create(curve), Write(func), run_time = 0.5)

    # Show limit equation to the right
    limit_eq = MathTex(
      R"""\approx \lim_{n\to\infty} \sum_{i=1}^{n} f(c_i) \Delta x = \int_{a}^{b} f(x) \,dx""",
      font_size = font_size
    )
    limit_eq.next_to(equation, RIGHT, buff = 0.5).shift(LEFT * 0.3)
    self.play(Write(limit_eq), run_time = 0.5)

    # Show 10 rectangles approximating the area
    self.play(Create(riemann_rects), run_time = 0.5)

    # Animate from 10 to 999 rectangles
    self.play(tracker.animate.set_value(250), run_time = 12)
    self.wait()
```
