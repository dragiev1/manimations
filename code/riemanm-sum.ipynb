```python
from manim import *
```

```python
class RiemannSums(Scene):
  def construct(self):
    equations = MathTex(
      R"""
        \begin{aligned}
        \sum_{i=1}^{n} f(c_i) \Delta x \leq \lim_{n\to\infty} \sum_{i=1}^{n} f(c_i) \Delta x &=
        \int_{a}^{b} f(x) \,dx \\
        \end{aligned}
      """
    )
    equations.to_corner(UR).scale(0.75)
    equations.shift(LEFT * 0.5)
    self.add(equations)
    axes = Axes(
      y_range=(0, 8, 1),
      x_range=(0, 8, 1),
      x_length=8,
      y_length=8,
      axis_config={"include_tip": False}
    )
    axes.center()
    axes.move_to(DL).shift(1.5 * UP + 0.5 * RIGHT)
    self.add(axes)
    self.play(Write(equations), run_time=2)
```

```python
manim -pql RiemannSums
```
